```python
from PIL import Image

def save_to_bmp(arr, img_name):
    min_value = arr.min()
    max_Value = arr.max()
    rescaled = (255.0 // (max_Value-min_value)  * (arr - min_value)).astype(np.uint8)
    img = Image.fromarray(rescaled)
    img.save(img_name, "BMP")
```

```python
for sub in subs:

    folder = 'Z:/Lisa/PARISK_bmp/' + sub + '/'

    t1w_slcs = []
    for item in glob.glob(folder + '/t1w*bmp'):
        t1w_slcs.append(item[-9:-4])

    t2w_slcs = []
    for item in glob.glob(folder + '/t2w*bmp'):
        t2w_slcs.append(item[-9:-4])

    tfe_slcs = []
    for item in glob.glob(folder + '/tfe*bmp'):
        tfe_slcs.append(item[-9:-4])

    tof_slcs = []
    for item in glob.glob(folder + '/tof*bmp'):
        tof_slcs.append(item[-9:-4])

    contrast_slcs = []
    for item in glob.glob(folder + '/contrast*bmp'):
        contrast_slcs.append(item[-9:-4])

    contour_slcs = []
    for item in glob.glob(folder + '/contour*bmp'):
        contour_slcs.append(item[-9:-4])

    slcs = list(set(t1w_slcs) & set(t2w_slcs) & set(tof_slcs) & set(tfe_slcs) & set(contour_slcs) & set(contrast_slcs))
    
    for sl in slcs:

        t1w_name = folder + 't1w' + sl + '.bmp'
        t2w_name = folder + 't2w' + sl + '.bmp'
        tof_name = folder + 'tof' + sl + '.bmp'
        tfe_name = folder + 'tfe' + sl + '.bmp'
        contrast_name = folder + 'contrast' + sl + '.bmp'
        contour_name = folder + 'contour' + sl + '.bmp'

        t1w = imageio.imread(t1w_name)[..., 0]
        t2w = imageio.imread(t2w_name)[..., 0]
        tof = imageio.imread(tof_name)[..., 0]
        tfe = imageio.imread(tfe_name)[..., 0]
        contrast = imageio.imread(contrast_name)[..., 0]
        contour = imageio.imread(contour_name)
        
        if (t1w.shape == t2w.shape) and (t1w.shape == tof.shape) and (t1w.shape == tfe.shape) and (t1w.shape == contrast.shape):
        
            dim = t1w.shape
            composed = np.zeros(dim)
            composed[:int(2*dim[0]/3), :int(dim[1]/2)] = t1w[:int(2*dim[0]/3), :int(dim[1]/2)]
            composed[int(2*dim[0]/3):, :int(dim[1]/2)] = t2w[int(2*dim[0]/3):, :int(dim[1]/2)]
            composed[:int(2*dim[0]/3), int(dim[1]/2):] = tof[:int(2*dim[0]/3), int(dim[1]/2):]
            composed[int(2*dim[0]/3):, int(dim[1]/2):] = tfe[int(2*dim[0]/3):, int(dim[1]/2):]
            composed[:int(dim[0]/3), ...] = contrast[:int(dim[0]/3), ...]

            xmin = 1
            xmax = dim[0]
            ymin = 1
            ymax = dim[1]

            if np.sum(contour) > 0:
                xmin = np.min(np.where(contour > 0)[0])-10
                xmax = np.max(np.where(contour > 0)[0])+15
                ymin = np.min(np.where(contour > 0)[1])-10
                ymax = np.max(np.where(contour > 0)[1])+15

            fig = plt.figure(figsize=(21, 14))

            plt.subplot(2, 3, 1)
            plt.imshow(t1w[xmin:xmax, ymin:ymax], cmap = 'gist_gray')
            if np.sum(contour) > 0:
                for i in np.unique(contour):
                    plt.contour((contour == i)[xmin:xmax, ymin:ymax], colors = ['yellow'], linewidths = 1)
            plt.title('t1w')
            plt.axis('off')

            plt.subplot(2, 3, 2)
            plt.imshow(t2w[xmin:xmax, ymin:ymax], cmap = 'gist_gray')
            if np.sum(contour) > 0:
                for i in np.unique(contour):
                    plt.contour((contour == i)[xmin:xmax, ymin:ymax], colors = ['yellow'], linewidths = 1)
            plt.title('t2w')
            plt.axis('off')

            plt.subplot(2, 3, 3)
            plt.imshow(tof[xmin:xmax, ymin:ymax], cmap = 'gist_gray')
            if np.sum(contour) > 0:
                for i in np.unique(contour):
                    plt.contour((contour == i)[xmin:xmax, ymin:ymax], colors = ['yellow'], linewidths = 1)
            plt.title('tof')
            plt.axis('off')

            plt.subplot(2, 3, 4)
            plt.imshow(tfe[xmin:xmax, ymin:ymax], cmap = 'gist_gray')
            if np.sum(contour) > 0:
                for i in np.unique(contour):
                    plt.contour((contour == i)[xmin:xmax, ymin:ymax], colors = ['yellow'], linewidths = 1)
            plt.title('tfe')
            plt.axis('off')

            plt.subplot(2, 3, 5)
            plt.imshow(contrast[xmin:xmax, ymin:ymax], cmap = 'gist_gray')
            if np.sum(contour) > 0:
                for i in np.unique(contour):
                    plt.contour((contour == i)[xmin:xmax, ymin:ymax], colors = ['yellow'], linewidths = 1)
            plt.title('contrast')
            plt.axis('off')

            plt.subplot(2, 3, 6)
            plt.imshow(composed)
            plt.title('coregistration check')
            plt.axis('off')

            #plt.show()   
            plt.savefig('Z:/Lisa/PARISK_check/' + sub + '_' + sl + '.png')
            plt.close()
            
        else:
            print(sub)
```
