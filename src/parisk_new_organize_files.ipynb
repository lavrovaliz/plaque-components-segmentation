```python
for item in os.listdir(d):
    path = os.path.join(d, item)
    if os.path.isdir(path):
        for subdir, dirs, files in os.walk(path):
            for filename in files:
                filepath = subdir + os.sep + filename
                dcm = pydicom.read_file(filepath, force = True)
                dcm.file_meta.TransferSyntaxUID = pydicom.uid.ImplicitVRLittleEndian
                try:
                    os.rename(filepath, path + os.sep + dcm.SOPInstanceUID + '.dcm')
                except:
                    print (filepath)
                #dcm.save_as(path + os.sep + dcm.SOPInstanceUID + '.dcm')
```

```python
for item in os.listdir(d):
    path = os.path.join(d, item)
    if os.path.isdir(path):
        for subitem in os.listdir(path):
            subpath = os.path.join(path, subitem)
            if os.path.isdir(subpath):
                shutil.rmtree(subpath)
```

```python
dcm_df = []

for item in os.listdir(d):
    path = os.path.join(d, item)
    if os.path.isdir(path):
        for subitem in os.listdir(path):
            subpath = os.path.join(path, subitem)
            dcm = pydicom.read_file(subpath, force = True)
            dcm.file_meta.TransferSyntaxUID = pydicom.uid.ImplicitVRLittleEndian
            rec = {'sub': item, 'series': dcm.SeriesInstanceUID, 'description': dcm.SeriesDescription}
            dcm_df.append(rec)
            
```

```python
dcm_df = pd.DataFrame(dcm_df)

dcm_df
```

```python
dcm_df['number_files'] = dcm_df['series'].map(dcm_df['series'].value_counts())
```

```python
dcm_df.drop_duplicates(subset=['sub', 'series'], keep='first', inplace = True)
```

```python
dcm_df
```

```python
series_df = []

for sub in np.unique(list(dcm_df['sub'])):
    temp_df = dcm_df.loc[dcm_df['sub']==sub].copy()
    rec = {'sub': sub, 'series': list(temp_df.description), 'n_files': list(temp_df.number_files)}
    series_df.append(rec)
    
series_df = pd.DataFrame(series_df)
```
